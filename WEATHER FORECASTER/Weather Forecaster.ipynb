```python
# Read Weather Dataset
# Recorded by the San Antonio International Airport, obtained through the NOAA
# Data is the weather of each day from Mar 1, 2017 to Feb 28, 2022
import pandas as pd
weather = pd.read_excel("WeatherDataset.xlsx")
weather.head()
```

```python
# Print Weather Dataset shape
weather.shape
```

```python
# Print the columns of the dataset
weather.columns
```

```python
# Replace the names of the data types (columns) as defined by their meaning given by the NOAA
weather.rename(columns = {"WSF2" : "Fastest 2-minute wind speed", 
                          "WSF5" : "Fastest 5-second wind speed",
                          "SNOW" : "Snowfall",
                          "WT03" : "Thunder",
                          "WT04" : "Ice pellets, sleet, snow pellets, or small hail",
                          "PRCP" : "Precipitation",
                          "WT05" : "Hail",
                          "WT06" : "Glaze or rime",
                          "WT07" : "Dust, volcanic ash, blowing dust/sand/obstruction",
                          "WT08" : "Smoke or haze",
                          "SNWD" : "Snow depth",
                          "WDF2" : "Direction of fastest 2-minute wind",
                          "AWND" : "Average wind speed",
                          "WDF5" : "Direction of fastest 5-second wind",
                          "PGTM" : "Peak gust time",
                          "WT01" : "Fog, ice fog, or freezing fog",
                          "TMAX" : "Maximum temperature",
                          "WT02" : "Heavy fog or heavy freezing fog",
                          "TAVG" : "Average temperature",
                          "TMIN" : "Minimum temperature"  
                         },
                inplace='True')
```

```python
# Drop STATION and NAME (redundant info)
weather.drop(columns=['STATION', 'NAME'])
```

```python
# Testing forecasting algorithm with Average temperature
# Setting DATE as the index
avgtemp = weather[["DATE", "Average temperature"]]
avgtemp = avgtemp.set_index('DATE')
avgtemp = avgtemp.asfreq('D')
avgtemp
```

```python
# Checking for missing values
print(f'Number of rows with missing values: {avgtemp.isnull().any(axis=1).mean()}')
```

```python
# First four of the five years will be used as training set (Mar 1, 2017 to Feb 28, 2021)
train = avgtemp[:'2021-02-28']
train
```

```python
# Data from the fifth year will be used as testing set (Mar 1, 2021 to Feb 28, 2022)
test = avgtemp['2021-03-01':]
test
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.rcParams['lines.linewidth'] = 1.5

%matplotlib inline

fig, ax=plt.subplots(figsize=(15, 5))
train['Average temperature'].plot(ax=ax, label='train')
test['Average temperature'].plot(ax=ax, label='test')
plt.title("Average Temperature (in degrees Fahrenheit)")
ax.legend();
```

```python
# Skforecast: allows the use of scikit-learn regressors as forecasters
# https://github.com/JoaquinAmatRodrigo/skforecast
from skforecast.ForecasterAutoreg import ForecasterAutoreg

# Importing the RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor

# Import the MSE (Mean Squared Error) as a metric
from sklearn.metrics import mean_squared_error

# Creating and training forecaster
forecaster = ForecasterAutoreg(regressor = RandomForestRegressor(random_state = 123),
                              lags = 365*2)

# Use squeeze() to convert the train DataFrame into a Series
forecaster.fit(y=train.squeeze())
```

```python
# Have forecaster predict the average temperature from Mar 1, 2021 to Feb 28, 2022
predictions = forecaster.predict(steps=365)

# Plot the predictions
fig, ax=plt.subplots(figsize=(15, 5))
train['Average temperature'].plot(ax=ax, label='train')
test['Average temperature'].plot(ax=ax, label='test')
predictions.plot(ax=ax, label='predictions')
plt.title("Average Temperature (in degrees Fahrenheit)")
ax.legend();
```

```python
# Prediction error
error_mse = mean_squared_error(y_true = test, y_pred = predictions)
print(f"Test error (mse): {error_mse}")
```

```python
# Predict average temperature for the next year (March 1, 2022 to Feb 28, 2023)
forecaster.fit(y=avgtemp.squeeze())

predictions = forecaster.predict(steps=365)
predictions
```
